```python
import csv
def read_data(filename):
    with open(filename, "r") as f:
        reader = csv.reader(f, delimiter=";")
        data = [row for row in reader]
    return data
read_data("student_cid.txt")
```

## Task 1.2

```python
def gender_count(cid_student_lst, is_female):
    if is_female:
        return len([i for i in cid_student_lst if i[2]=="F"])
    else:
        return len([i for i in cid_student_lst if i[2]=="M"])

cid_student_lst = read_data("student_cid.txt")
print(gender_count(cid_student_lst, True))
print(gender_count(cid_student_lst, False))
```

## Task 1.3

```python
def role_statistics(cid_student_lst):
    role_num = {i[1]:0 for i in cid_student_lst}
    for i in cid_student_lst:
        role_num[i[1]] += 1
    print("Role\t\tNumber")
    for i in role_num:
        if len(i) < 8: # pretty print
            print(f"{i}\t\t{role_num[i]}")
        else:
            print(f"{i}\t{role_num[i]}")

cid_student_lst = read_data("student_cid.txt")
role_statistics(cid_student_lst)
```

## Task 1.4

```python
import random
def form_random_group(cid_student_lst):
    roles = ["Coder", "Maker", "Dealer", "Empathizer", "Designer"]
    students = [ [j[0] for j in cid_student_lst if j[1]==i] for i in roles ]
    
    random_grp = []
    for i in students:
        if len(i) > 0:
            random_grp.append(i[random.randint(0,len(i)-1)])
        else:
            break
    
    return random_grp

cid_student_lst = read_data("student_cid.txt")
form_random_group(cid_student_lst)
```

## Task 1.5

```python
def remove_students(cid_student_lst, one_cid_group):
    removed_records = []
    for i in cid_student_lst:
        if i[0] in one_cid_group:
            removed_records.append(i)
    for i in removed_records:
        cid_student_lst.remove(i) # this actually modifies cid_student_lst since it is mutable
    return removed_records

cid_student_lst = read_data("student_cid.txt")
one_cid_group = form_random_group(cid_student_lst)
removed_records = remove_students(cid_student_lst, one_cid_group)
print("removed records")
for removed_record in removed_records:
    print(removed_record)
print("remaining records")
print(len(cid_student_lst))
for cid_student in cid_student_lst:
    print(cid_student)
```

## Task 1.6

```python
def form_max_cid_group(cid_student_lst):
    groups = []
    while True:
        one_cid_group = form_random_group(cid_student_lst)
        if one_cid_group == []:
            break
        removed_students = remove_students(cid_student_lst, one_cid_group)
        groups.append(removed_students)
    
    with open("result.txt", "w") as f:
        for i in range(len(groups)):
            f.write(f"Group {i}\n")
            for j in groups[i]:
                f.write(f"{j[0]} {j[1]} {j[2]}\n")

cid_student_lst = read_data("student_cid.txt")
form_max_cid_group(cid_student_lst)
```

```python
def EAN(ean12):
    odd_digits = 0
    even_digits = 0
    for i in range(len(ean12)):
        if i % 2 == 0:
            even_digits += int(ean12[i])
        else:
            odd_digits += int(ean12[i])
    total = even_digits + odd_digits*3
    check_digit = ( 10 - int(str(total)[-1]) )%10 # so that if total=10, check_digit=0 instead of 10
    return f"{ean12}{check_digit}"

print(EAN("400638133393"))
print(EAN("590123412345"))
print(EAN("950110153000"))
print(EAN("007567816412"))
print(EAN("123456789123"))
print(EAN("563643712973"))
```

## Task 2.2

```python
def EAN_rec_main(ean12, counter, total):
    if len(ean12) == 0:
        return ( 10 - int(str(total)[-1]) )%10
    if counter % 2 == 0:
        return EAN_rec_main(ean12[1:], counter+1, total+int(ean12[0]))
    else:
        return EAN_rec_main(ean12[1:], counter+1, total+(int(ean12[0]))*3)

def EAN_rec(ean12):
    return f"{ean12}{EAN_rec_main(ean12,0,0)}"

print(EAN_rec("400638133393"))
print(EAN_rec("590123412345"))
print(EAN_rec("950110153000"))
print(EAN_rec("007567816412"))
print(EAN_rec("123456789123"))
print(EAN_rec("563643712973"))
```

```python
import random
def generate_n_random_EAN(n):
    twelve_digit_nums = []
    for i in range(n):
        s = ""
        for j in range(12):
            s = s + str(random.randint(0,9))
        twelve_digit_nums.append(s)

    return [EAN(i) for i in twelve_digit_nums]

generate_n_random_EAN(5)
```

## Task 2.4

```python
def qs(array): # not in-place
    if len(array) == 0:
        return []
    mid = (0+len(array))//2
    pivot = array[mid]
    lesser = [i for i in array if i < pivot]
    equal = [i for i in array if i == pivot]
    greater = [i for i in array if i > pivot]
    return qs(lesser) + equal + qs(greater)

def quick_sort_10_EAN():
    eans = generate_n_random_EAN(10)
    return qs(eans)

quick_sort_10_EAN()
```
